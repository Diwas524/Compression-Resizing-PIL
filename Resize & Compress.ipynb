```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
 
image = cv2.imread("ds.png", 1)
# Loading the image
 

 
stretch_near = cv2.resize(image, (780, 540),
               interpolation = cv2.INTER_NEAREST)
 
cv2.imwrite("resize.jpg",stretch_near)
Titles =["Original", "Interpolation Nearest"]
images =[image,stretch_near]
count = 2
 
for i in range(count):
    plt.subplot(2, 2, i + 1)
    plt.title(Titles[i])
    plt.imshow(images[i])
 
plt.show()
```

Resize & Compress using PIL

```python
from PIL import Image

img = Image.open("resize.jpg")
print(img.size)

img = img.resize((750,550),Image.ANTIALIAS)  #resizing

img.save("image_scaled.jpg",quality=95,optimize=True)  #compressing; you can change quality as needed
```
